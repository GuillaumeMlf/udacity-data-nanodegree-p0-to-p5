```python
import pandas as pd
import numpy as np
import pprint as pp
pd.options.display.max_columns= 999
pd.options.display.max_rows= 999
```

```python
data_dict = pd.read_csv("pisadict2012.csv")
data = pd.read_csv("pisa2012.csv")
```

Dataset of 485490 observations and 636 columns.
France: 4613 obs. 
Non-OECD: 190074 obs.
OECD: 295416 obs.

PV1 MATH AVE = PV1MATH*FINAL_STUDENT_WEIGHT/SUM(FINAL_STUDENT_WEIGHT)
PV2 MATH AVE = PV2MATH*FINAL_STUDENT_WEIGHT/SUM(FINAL_STUDENT_WEIGHT)
PV3 MATH AVE = PV3MATH*FINAL_STUDENT_WEIGHT/SUM(FINAL_STUDENT_WEIGHT)
PV4 MATH AVE = PV4MATH*FINAL_STUDENT_WEIGHT/SUM(FINAL_STUDENT_WEIGHT)
PV5 MATH AVE = PV5MATH*FINAL_STUDENT_WEIGHT/SUM(FINAL_STUDENT_WEIGHT)

MATH AVE = (PV1 MATH AVE + PV2 MATH AVE + PV3 MATH AVE + PV4 MATH AVE + PV4 MATH AVE + PV5 MATH AVE) / 5


```python
data['math_ave'] = (((data['PV1MATH'] * data['W_FSTUWT']) / np.mean(data['W_FSTUWT'])) +
                    ((data['PV2MATH'] * data['W_FSTUWT']) / np.mean(data['W_FSTUWT'])) +
                    ((data['PV3MATH'] * data['W_FSTUWT']) / np.mean(data['W_FSTUWT'])) +
                    ((data['PV4MATH'] * data['W_FSTUWT']) / np.mean(data['W_FSTUWT'])) +
                    ((data['PV5MATH'] * data['W_FSTUWT']) / np.mean(data['W_FSTUWT']))) / 5
```

Analysis will focus on:
    ST88Q02 = Waste of Time
    ST86Q05 = Teachers Treat Students Fair

```python
df_strongly_agree = data[data['ST86Q05'] == 'Strongly agree'].groupby(['OECD','ST88Q02'])['math_ave'].agg(['mean', 'count'])
df_agree = data[data['ST86Q05'] == 'Agree'].groupby(['OECD','ST88Q02'])['math_ave'].agg(['mean', 'count'])
df_disagree = data[data['ST86Q05'] == 'Disagree'].groupby(['OECD','ST88Q02'])['math_ave'].agg(['mean', 'count'])
df_strongly_disagree = data[data['ST86Q05'] == 'Strongly disagree'].groupby(['OECD','ST88Q02'])['math_ave'].agg(['mean', 'count'])
```

```python
df_strongly_disagree['Treat Students Fair'] = 'Strongly disagree'
df_disagree['Treat Students Fair'] = 'Disagree'
df_agree['Treat Students Fair'] = 'Agree'
df_strongly_agree['Treat Students Fair'] = 'Strongly agree'
```

```python
result = pd.concat([df_strongly_agree, df_agree, df_disagree, df_strongly_disagree])
```

```python
result = result.reset_index(level=['OECD', 'ST88Q02'])
result.rename(columns={'ST88Q02':'Attitue towards School - Waste of time'}, inplace=True)
result.rename(columns={'ST88Q02':'Attitue towards School - Waste of time'}, inplace=True)
result.rename(columns={'mean':'Math score'}, inplace=True)
```
